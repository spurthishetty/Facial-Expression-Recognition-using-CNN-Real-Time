These works differ significantly in terms of CNN architectures

We have also used methods to reduce overfitting thus increasing the performance of the model on unknown data. Next, let's open the csv file and try to count the number of samples we have in our file:  
This simple code counts the number of samples and along with that it gives the instance length. An instance length is basically a value of the number of pixels in each of the pixels column. The image size is 48x48 thus the instance length is 48 x 48 = 2304

```python
# Count number of samples and sample size 

with open("./fer2013.csv") as f:
    content = f.readlines()

lines = np.array(content)

num_of_instances = lines.size
print("number of instances: ",num_of_instances)
print("instance length: ",len(lines[1].split(",")[1].split(" ")))
```

After counting the number of samples, let's sort into train, test and validation sets. Below, we have made lists of training, testing and validation sets:

```python
#Initialize train, test and validation set

x_train, y_train, x_test, y_test, x_val, y_val = [], [], [], [], [], []
```

We have now initialized a for loop to sort all samples for our data.  
The emotion and pixel values with usage as "Training" will be assigned to x_train and Y-train sets respectively.  
Similarly, emotion and pixel values with usage as "PublicTest" will be assigned to x_test and Y-test sets respectively.  
Lastly, emotion and pixel values with usage as "PrivateTest" will be assigned to x_val and Y-val sets respectively.
* https://www.tensorflow.org/api_docs/python/tf/keras/utils/to_categorical

```python
#Transfer data into these sets

for i in range(1,num_of_instances):
    try:
        emotion, img, usage = lines[i].split(",")
          
        val = img.split(" ")
            
        pixels = np.array(val, 'float32')
        
        emotion = keras.utils.to_categorical(emotion, num_classes)
    
        if 'Training' in usage:
            y_train.append(emotion)
            x_train.append(pixels)
        elif 'PublicTest' in usage:
            y_test.append(emotion)
            x_test.append(pixels)
        elif 'PrivateTest' in usage:
            y_val.append(emotion)
            x_val.append(pixels)
    except:
        print("",end="")
```

After sorting the data into train, test and validation sets we need to convert them to numpy arrays for processing and later use them for training, validation and testing purposes.  

Hence, first we try to convert each list into numpy array after that we normalize the input between 0 and 1.  
After these steps we reshape the array to the size of pur input. Input is infact our image and the size of our image is 48x48 so  thus we reshape the input samples to a shape of (48, 48, 1). Here 1 is the number of channels of an image. We are working with grayscale images and these images have 1 channel hence we the value of 1 is mentioned.  



Later we have caluclated the number of samples in each of the set. The percentage split for training, testing and validation is 80%, 10% and 10% respectively.
* https://machinelearningmastery.com/why-one-hot-encode-data-in-machine-learning/

```python
#Data transformation for train, test and validation sets

x_train = np.array(x_train, 'float32')
y_train = np.array(y_train, 'float32')
x_test = np.array(x_test, 'float32')
y_test = np.array(y_test, 'float32')
x_val = np.array(x_val, 'float32')
y_val = np.array(y_val, 'float32')

#normalize inputs between [0, 1]

x_train /= 255 
x_val /= 255
x_test /= 255

x_train = x_train.reshape(x_train.shape[0], 48, 48, 1)
x_train = x_train.astype('float32')
x_val = x_val.reshape(x_val.shape[0], 48, 48, 1)
x_val = x_val.astype('float32')
x_test = x_test.reshape(x_test.shape[0], 48, 48, 1)
x_test = x_test.astype('float32')

#Count the number of samples in each set

print(x_train.shape[0], 'train samples')
print(x_test.shape[0], 'test samples')
print(x_val.shape[0], 'validaiton samples')
```

Our main aim was to build a simple network from which we can take out maximum accuracy and thus make the correct predictions. We have used 3 pairs of convolution layers with an input batch of size 64 and as the number of layers increases, the batch size also increases. We have taken a kernel size of (5,5) in the input layer and thus it is een that we have decreased the size to (3,3) in the further layers to extract maximum information from the input.  
The activation function we have used here is RELU which stands for Rectified Linear Unit.  

In the beginning iterations of increasing the accracy of our model, the model often used to overfit. Increasing the number of epochs also didn;t do much good as the training accuracy would increase gradually but the testing accracy was stagnant at 57% back then. After sometime, we came across the methods of data augmentation and batch normaliztion and we have successfully to implemente them in our model. The result is an increase in accuracy by 6-7%. Thus these are proven methods to reduce overfittting anf thus improve accuracy. So here we have added a layer of batch normalization just after the convolution layer and max pooling layer after that. Amongst all the pooling types we choose max pooling as it is proven to provide the  best reuslts. The pool size is a 2x2 matrix to collect up features.  

At the end we have used a couple of fully connected layers. Each layer here is followed by a layer of batch normalization, activation and dropout. Drop out randomnly drops specified amount of samples to fight overfitting.  

At the end we have added a dense layer to classify various inputs in the appropriate emotion.

```python
#Building the network

model = Sequential()

#1st Convolution layer
model.add(Conv2D(64, (5, 5), activation='relu', input_shape=(48,48,1)))
model.add(Conv2D(64, (5, 5), activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2,2), strides=(2, 2)))

#2nd Convolution layer
model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2,2), strides=(2, 2)))

#3rd Convolution layer
model.add(Conv2D(256, (3, 3), activation='relu'))
model.add(Conv2D(256, (3, 3), activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2,2), strides=(2, 2)))

model.add(Flatten())

#Fully Connected Layers

model.add(Dense(1024))#, activation='relu'))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(Dropout(0.3))
model.add(Dense(1024))#, activation='relu'))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(Dropout(0.3))

model.add(Dense(num_classes, activation='softmax'))

model.summary()
```

Below is the process of data augmentation where new samples are generated for the model to learn. Tweaking and changing various parameters here generates variations of the same sample images that we have in our dataset thus creating more data for our model to learn and thus this is a way to reduce overfitting. 


```python
#Data Augmentation and Batch Process


gen = ImageDataGenerator(
      rotation_range = 45,
      width_shift_range = 0.2,
      height_shift_range = 0.2,
      shear_range = 0.2,
      zoom_range = 0.2,
      horizontal_flip=True,
      fill_mode = 'nearest')


train_generator = gen.flow(x_train, y_train, batch_size=batch_size)
```

Now we have come to the compilation step where we compile the model alongside defining the loss function which lets us measure how good a prediction model does in terms of being able to predict the expected outcome and for that we have used categorical_crossentropy which is function usually used when you have to make a prediction of n number of classes.  

Next is the optimizer. We ended up using Adam as our optimal optimizer after trying various different optimizers from the keras library. It is said to be the best amongst all of the other optimizers which provides an ideal optimization by changing the necessary weights of the model. We have kept the learning rate a standard of 0.0001. Lastly, we have used a metrics of "accuracy" which will be using further in our notebook.

```python
#The compilation step

model.compile(loss='categorical_crossentropy'
    , optimizer=keras.optimizers.Adam(lr = 0.0001)
    , metrics=['accuracy'])
```

Finally after a lot of data preprocessing steps, we have come to the most important part of the notebook that is the training of our model. Here we fit our model with the inputs as the from the train_generator, we also make use of the validation data to simultaneously validate our data against that dataset to make sure that our model isn't overfitting as we progress further or increase the number of epochs. Previously, we have defined the batch size and the number of epochs so we are going to use that now.

```python
#Fit the model

history = model.fit_generator(train_generator, 
                              validation_data = (x_val, y_val), 
                              steps_per_epoch=batch_size, 
                              epochs=epochs)
```

This is a wonderful piece of code we found online in a GitHub repository of Francois Chollet, who is the creator of Keras. The link to the code is right here: https://github.com/fchollet/deep-learning-with-python-notebooks/blob/master/5.4-visualizing-what-convnets-learn.ipynb  

He had visualized the first three of layers of his CNN model in a very intuative and eye catching format. We made changes in this code, passed our own sample image into it and visualized all the 12 layers of our model.  

As you can see the first image depicts an array of 16 channels of the first layer the input image. The possible image of a human can be identified in this layer. In the next layer, the image starts getting fuzzier and the picture starts to fade out as we go deeper into the layers. Why does that happen? As a result of the convolution step, features are extracted from the input layers by the filters to form different feature maps. In the max pooling step, the best features from these feature maps are extracted and the clarity or the identifying property of the content in the image is lost and we cannot identify the contents of the image. This process takes place in each and every convolution and max pooling step.  

One more property to notice is that, the size of the layer images increases as we go further. This is because the size of feature maps begins to increase once we go deeper into the layers.

```python

# These are the names of the layers, so can have them as part of our plot
layer_names = []
for layer in model.layers[:12]:
    layer_names.append(layer.name)

images_per_row = 16

# Now let's display our feature maps
for layer_name, layer_activation in zip(layer_names, activations):
    # This is the number of features in the feature map
    n_features = layer_activation.shape[-1]

    # The feature map has shape (1, size, size, n_features)
    size = layer_activation.shape[1]

    # We will tile the activation channels in this matrix
    n_cols = n_features // images_per_row
    display_grid = np.zeros((size * n_cols, images_per_row * size))

    # We'll tile each filter into this big horizontal grid
    for col in range(n_cols):
        for row in range(images_per_row):
            channel_image = layer_activation[0,
                                             :, :,
                                             col * images_per_row + row]
            # Post-process the feature to make it visually palatable
            channel_image -= channel_image.mean()
            channel_image /= channel_image.std()
            channel_image *= 64
            channel_image += 128
            channel_image = np.clip(channel_image, 0, 255).astype('uint8')
            display_grid[col * size : (col + 1) * size,
                         row * size : (row + 1) * size] = channel_image

    # Display the grid
    scale = 1. / size
    plt.figure(figsize=(scale * display_grid.shape[1],
                        scale * display_grid.shape[0]))
    plt.title(layer_name)
    plt.grid(False)
    plt.imshow(display_grid, aspect='auto', cmap='viridis')
    
plt.show()
```

```python
emotions = ('angry', 'disgust', 'fear', 'happy', 'sad', 'surprise', 'neutral')

while(True):
    ret, img = cap.read()
    #img = cv2.imread('C:/Users/IS96273/Desktop/hababam.jpg')

    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)

    #print(faces) #locations of detected faces

    for (x,y,w,h) in faces:
        cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2) #draw rectangle to main image
        
        detected_face = img[int(y):int(y+h), int(x):int(x+w)] #crop detected face
        detected_face = cv2.cvtColor(detected_face, cv2.COLOR_BGR2GRAY) #transform to gray scale
        detected_face = cv2.resize(detected_face, (48, 48)) #resize to 48x48
        
        img_pixels = image.img_to_array(detected_face)
        img_pixels = np.expand_dims(img_pixels, axis = 0)
        
        img_pixels /= 255 #pixels are in scale of [0, 255]. normalize all pixels in scale of [0, 1]
        
        predictions = model.predict(img_pixels) #store probabilities of 7 expressions
        
        #find max indexed array 0: angry, 1:disgust, 2:fear, 3:happy, 4:sad, 5:surprise, 6:neutral
        max_index = np.argmax(predictions[0])
        
        emotion = emotions[max_index]
        
        #write emotion text above rectangle
        cv2.putText(img, emotion, (int(x), int(y)), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,255,255), 2)
        
        #process on detected face end
        #-------------------------
        cv2.imshow('img',img)

    if cv2.waitKey(1) & 0xFF == ord('q'): #press q to quit
        break

#kill open cv things
cap.release()
cv2.destroyAllWindows()
```
